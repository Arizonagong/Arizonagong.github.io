2. Calculating normal distribution

To calculate probability of specific value to happen, we need to identify

1. Mean: m
2. Standard deviation: sd
3. Number of samples: n

Greer will guarantee that its engines will provide a certain amount of lifetime flight hours 
subject to the airline purchasing a preventive-maintenance service plan that is also provided by Greer.

With proper maintenance service, their engine’s lifetime flight hours achieved is **normally
distributed**, **mean=36500**, **SD=5000**.

Question1. They want to promote that once they purchase the maintenance service pack, then you don’t
need to replace your engine until you use it for 40,000 hrs by XX%

So, Greer would like to know what percentage of its aircraft engines will be expected to last
more than 40,000hrs. That is, what is the probability that the aircraft lifetime flight hours x
will exceed 40,000?

```python
# Create data: Assuming randomly selected and normally distributed.
m=40000
sd=6000
n=1000000
f=50000

# loc=mean, scale=standard deviation
value= np.random.normal(loc=m, scale=sd, size=n)
value
points=sns.kdeplot(value, fill=False, bw_method=1.0, color='black').get_lines()[0].get_data()
cum=stat.NormalDist(mu=m, sigma=sd).cdf(f)
cum
print("A probability that the aircraft lifetime flight hours will exceed", f, "hrs is", 
      1-cum,"%.")
```
