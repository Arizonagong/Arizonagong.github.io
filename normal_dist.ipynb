# 2. Calculating Normal Distribution

## (1) Creating normal distribution

To calculate probability of specific value to happen, we need to identify:
1. **Mean**: m
2. **Standard Deviation**: sd
3. **Number of Samples**: n

Greer will guarantee that its engines will provide a certain amount of lifetime flight hours subject to the airline purchasing a preventive-maintenance service plan that is also provided by Greer.

With proper maintenance service, their engineâ€™s lifetime flight hours achieved is normally
distributed, mean=36500, SD=5000.

```python
# Create data: Assuming randomly selected and normaly distributed
m=87
sd=6.5
n=25

# Create random data that can meet the condition of m, sd, and n.
value = np.random.normal(loc=m,scale=sd,size=n)

# Draw a normal distribution plot with 'value' data
points = sns.kdeplot(value, fill=False, bw_method=1.0, color='black').get_lines()[0].get_data()

# Draw a vertical line indicating mean value
plt.axvline(x=m)
```

### Question1. 

They want to promote that once they purchase the maintenance service pack, then you donâ€™t need to replace your engine until you use it for 40,000 hrs by XX%.

So, Greer would like to know what percentage of its aircraft engines will be expected to last more than 40,000hrs. That is, what is the probability that the aircraft lifetime flight hours x will exceed 40,000.

```python
f=90
# Calculating a probability that the aircraft lifetime flight hours x will exceed 40,000.
cum=stat.NormalDist(mu=m, sigma=sd).cdf(f)
print("A probability that the aircraft lifetime flight hours will exceed", f, "is", round(1-cum,4))
```

```python
x=points[0]
y=points[1]
```

```python
plt.fill_between(x, y, color='y')
plt.fill_between(x, y, where = (x>=f), color='r')
```

### Question2. 

Now Greer is considering a guarantee that will provide a discount on a replacement aircraft engine if the original engine does not meet the lifetime-flight-hour guarantee.

How many lifetime flight hours should Greer guarantee if Greer wants no more than 10% of aircraft engines to be eligible for the discount guarantee?

```python
inv=stat.NormalDist(mu=m, sigma=sd).inv_cdf(0.25)
print("Lifetime flight hours that can guarantee that no more than 10% of its engines will be eligible for the discount promotion is", round(inv,0),"flight hours.")
inv
```

```python
plt.fill_between(x, y, color='y')
plt.fill_between(x, y, where = (x<=inv), color='r')
```
