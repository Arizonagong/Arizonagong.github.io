```python
# "pip install scikit-learn" in your terminal to install the library
# "pip install seaborn"
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error,r2_score

import pandas as pd
```

```python
# Load your mtcars data
mtcars = pd.read_csv("mtcars.csv")
# mpg: miles per gallon
# cyl: number of cylinders
# disp: Displacement
# hp : horsepower
# drat: rear axie ratio
# wt: weight
#qsec: 1/4 mile time
# vs: engine type(0=V-shape, 1=strate)
# am: trasmition(0=automatic, 1=manual)
# gear: number of forward gears
```

```python
# Create plot
sns.lmplot(x='mpg',y='wt',data=mtcars, aspect=2, height=6)
```

```python
# Multiple Linear Regression
# Set your independent variables
x=mtcars[['hp','drat','disp']]
# Set your dependent variable
y=mtcars['mpg']

# Create your regression object
regr=LinearRegression()

# Fit your model(create your model)
regr.fit(x,y)
r_2=regr.score(x,y)

# Print result
print("Intercept:", regr.intercept_)
print("Coefficients:",list(zip(x, regr.coef_)))
print("R2:",r_2)
```
