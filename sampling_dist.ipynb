```python
# Sample data 5 times with a size of 5
# sample size = 5
# sampling repetition = 5

# size of sample
n=50
# sampling repetition number
rep=100

lst=[]
for _ in range(n):
    df_sampl=df.sample(rep) # Do random sampling of size 5
    sample_m=df_sampl["annual_salary"].mean()
    
    # Calculate the mean of the sample
    lst.append([sample_m]) # Add the calculated mean into the list

df_new = pd.DataFrame(lst, columns=["sample_mean"])
print(df_new)
df['annual_salary'].mean()
df_new.plot.kde()
```
