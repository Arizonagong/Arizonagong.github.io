```python
import pandas as pd
```

```python
df = pd.read_csv('EAI.csv')
print(df)
type(df)
df.describe()
df.columns = ["manager", "annual_salary","training_program"]
print(df)
```

```python
df.sample(10)
```

```python
df.sample(frac=0.1)
```

```python


lst=[]
for _ in range(5):
    df_sampl=df.query("training_program=='No'").sample(n=5)
    sample_m=df_sampl["annual_salary"].mean()
    lst.append([sample_m])
    
df_new = pd.DataFrame(lst, columns=["sample_mean"])
print(df_new)

df_new.plot.kde()

```
