```python
# "pip install scikit-learn" in your terminal to install the library
# "pip install seaborn"
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error,r2_score
```

1. Simple Linear Regression

```python
# Load the dataset: diabetes
diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)
# Use only one feature(independent variable)
diabetes_X = diabetes_X[:,np.newaxis,2]
```

```python
# Create linear regression object
regr=linear_model.LinearRegression()

# Fit the model(create the simple linear regression model)
regr.fit(diabetes_X, diabetes_y)

# Create a predicted value out of your model
diabetes_y_pred = regr.predict(diabetes_X)

# Calculate R square
r_sq = regr.score(diabetes_X, diabetes_y)
```

```python
# The slope value = Regression Coefficients
print("Coefficients:", regr.coef_)

# The R square value
print("R square:", r_sq)
```

```python
# Plot the simple linear regression
plt.scatter(diabetes_X,diabetes_y, color="black")
plt.plot(diabetes_X,diabetes_y_pred, color="blue", linewidth=3)
```

```python
# seaborn library 
sns.regplot(x=diabetes_X, y=diabetes_y)
```
